```python
# Machine Tranlation using IBM model 1

# Input: english sentence e, foreign sentence f, hash of translation probabilities t, epsilon 
# Output: probability of e given f

# Step 1: # Extract foreign and english vocabularies

# Step 2: # Alignment 
```

```python
import pandas as pd

df_eng_sents  = pd.read_csv('English_1.csv')['Text'].apply(lambda x: x.split())

df_fr_sents  = pd.read_csv('French_1.csv')['Text'].apply(lambda x: x.split())
```

```python
eng_sents_all = [[y.lower() for y in x] for x in df_eng_sents]

fr_sents_all = [[y.lower() for y in x] for x in df_fr_sents]
```

```python
# Sample 

print(eng_sents_all[:3])
```

```python
print(fr_sents_all[:3])
```

```python
import pandas as pd
from nltk.translate.ibm1 import IBMModel1
from nltk.translate.api import AlignedSent

# Alignment : Return an aligned sentence object, which encapsulates two sentences along with an Alignment between them.

max_count=500

eng_sents = eng_sents_all[:max_count]
fr_sents = fr_sents_all[:max_count]

print("Size of english sentences: ", len(eng_sents))
print("Size of french sentences: ", len(fr_sents))

aligned_text = []

for i in range(len(eng_sents)):
    al_sent = AlignedSent(fr_sents[i],eng_sents[i])
    aligned_text.append(al_sent)

aligned_text
```

```python
print("Training smt model")
ibm_model = IBMModel1(aligned_text,5)
print("Training complete")
```

```python
import random

n_random = random.randint(0,99)

print(n_random)

fr_sent = fr_sents[n_random]

print(fr_sent)

eng_sent_actual_tr = eng_sents[n_random]

tr_sent = []

for w in fr_sent:
    probs = ibm_model.translation_table[w]
    
    if(len(probs)==0):
        continue
    
    sorted_words = sorted([(k,v) for k, v in probs.items()],key=lambda x: x[1], reverse=True)
    top_word = sorted_words[1][0]
    
    if top_word is not None:
        tr_sent.append(top_word)

print("French sentence: ", " ".join(fr_sent))
print("Translated Eng sentence: ", " ".join(tr_sent))
print("Original translation: ", " ".join(eng_sent_actual_tr))
```
