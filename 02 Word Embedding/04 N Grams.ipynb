```python
# N-Grams as Features

# A combination of N words together are called N-Grams. N grams (N > 1) are generally more informative as 
# compared to words (Unigrams) as features. 
# Also, bigrams (N = 2) are considered as the most important features of all the others.


# Collocations: Bigrams and Trigrams

# Collocations are the pairs of words occurring together many times in a document. 
# It is calculated by the number of those pair occurring together to the overall word count of the document.

# Collocation can be categorized into two types-

# Bigrams combination of two words
# Trigramscombinationof three words
```

### N Grams Manual

```python
data = "Hi how are you"

data = data.split()

print(data)

ngram =2

# case 1: N Gram generator

for i in range(len(data) - (ngram -1)):
    
    start = i
    end = ngram + i
    print(data[start : end])
    
    
# case 2: collect Input and output data

ngram = 2
for i in range(len(data) - (ngram)):
    start = i
    end = ngram + i
    
    input = data[start : end]
    output = data[end]
    print("Input = {0} and output ={1}".format(input,output))
    
```

```python

def generate_ngrams(text,n):
    words = text.split()
    output= []
    for i in range(len(words) - n+1):
        output.append(words[i:i+n])
    return output

generate_ngrams('this is a sample text', 2)
```

### N Gram Using NLTK (bigrams)

```python
import nltk

text = "Guru99 is a totally new kind of learning experience."
Tokens = nltk.word_tokenize(text)
output = list(nltk.bigrams(Tokens))
print(output)
```

### N Gram Using NLTK (trigrams)

```python
import nltk
text = "Guru99 is a totally new kind of learning experience."
Tokens = nltk.word_tokenize(text)
output = list(nltk.trigrams(Tokens))
print(output)
```
