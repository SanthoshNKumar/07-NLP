### Simple List

```python
def one_hot_encode_my_word(myword,the_vocabulary):
    one_hot_encode_vector = []
    for word in the_vocabulary:
        if word == myword:
            one_hot_encode_vector.append(1)
        else:
            one_hot_encode_vector.append(0)
    return one_hot_encode_vector

v = ["duct", "tape", "works", "anywhere", "is", "magic", "and", "should", "be", "worshiped"]

for i in range(0,len(v)):
    print(v[i] + ": \t" +str(one_hot_encode_my_word(v[i],v)))
```

```python
import pandas as pd

train = [("Thanks for an excellent report", "pos"),
         ("Your service is very quick and fast", "pos"),
        ("I am pleased with your service", "pos"),
        ("I did not know i was diabetic until you gave me this report", "neg"),
        ("Service - Little slow, probably because too many people.", "neg"),
        ("The place is not easy to locate", "neg"),
        ("The place is very easy to locate", "pos"),
        ("Not satisfied will take a second opinion", "neg"),
        ("No human contact everything is so robotic here", "neg")]

df = pd.DataFrame(train,columns=['review','sentiment'])

df.head()


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stopwords = stopwords.words('english')

data = list(df['review'].values)

data

data_token = [word_tokenize(x.lower()) for x in data]
clean_data = []

for sent in data_token:
    #print([x for x in sent if (x not in stopwords and x not in "-.,")])
    clean_data.append([x for x in sent if (x not in stopwords and x not in "-.,")])

clean_data
```

```python
wrds = []

[[wrds.append(w.lower()) for w in x] for x in clean_data]

wrds = set(wrds)

wrds
```

```python
word_id = {}

for i,w in enumerate(wrds):
    word_id[w] = i

word_id
```

```python
row = len(clean_data)
print("row:",row)

vocab_size = len(word_id)

print("vocab_size=",vocab_size)

import numpy as np
onehot = np.zeros((row,26))

for i in range(len(clean_data)):
    doc = clean_data[i]
    print(doc)
    for j in range(len(doc)):
        id = word_id[(doc[j]).lower()]
        print('word = {0} and id = {1}'.format(doc[j],id))
        
        onehot[i][id] = 1
        
onehot
```

```python
import pandas as pd

pd.DataFrame(onehot,columns=wrds)
```
